```python
import numpy as np
import pandas as pd
from datetime import datetime
%matplotlib inline
```

```python
def parse_timestamp(x):
    date, time = x.split(" ")
    dates = date.split("/")
    times = time.split(":")
    return pd.to_datetime(datetime(int(dates[2]), int(dates[1]), int(dates[0]), int(times[0]), int(times[1])))
    
    
df = pd.read_csv("data/coffees.csv")
df.loc[2, "Coffees"] = np.nan
df["timestamp"] = df.Timestamp.apply(parse_timestamp)

# Random dropout
idx = list(np.random.randint(0, len(df), 12))
for i in idx:
    df.loc[i, "Coffees"] = np.nan
```

```python
df = df[["timestamp", "Coffees", "Name"]]
df.Coffees = df.Coffees.apply(lambda x: int(x) if np.isfinite(x) else np.nan)
```

```python
df.Name = df.Name.apply(lambda x: "Anthony" if x == "Ant" else x)
df.loc[2, "Coffees"] = "439."
df.sort_values("timestamp")
```

```python
from datetime import timedelta, date
boop = pd.DataFrame({"date": pd.date_range(df.loc[0, "timestamp"] - timedelta(days=17), "2013-05-10")})
boop["status"] = "OK"
boop.index = pd.to_datetime(boop.date)
boop.loc["2011-12-13":"2011-12-19", "status"] = "BROKEN"
boop.loc["2012-03-12":"2012-04-04", "status"] = "BROKEN"
boop.loc["2012-04-11":"2012-04-26", "status"] = "BROKEN"
boop.loc["2012-12-21":"2013-01-08", "status"] = "BROKEN"
```

```python
boop[["date", "status"]].to_csv("coffee_status.csv")
```
